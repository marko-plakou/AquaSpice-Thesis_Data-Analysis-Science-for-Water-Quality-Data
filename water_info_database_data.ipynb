```python
import pandas as pd
import numpy as np
```

## Read Data

```python


Dendermonde_pH=pd.read_csv('Dendermonde SF Dender_pH.csv',delimiter=';',header=7)
Dendermonde_Watertemperature=pd.read_csv('Dendermonde SF Dender_Watertemperature.csv',delimiter=';',header=7)
Dendermonde_Turbidity=pd.read_csv('Dendermonde SF Dender_Turbidity.csv',delimiter=';',header=7)
Dendermonde_Dissolved_Oxygen=pd.read_csv('Dendermonde SF Dender_Dissolved Oxygen.csv',delimiter=';',header=7)
Dendermonde_Conductivity=pd.read_csv('Dendermonde SF Dender_Conductivity.csv',delimiter=';',header=7)

```

```python
Dendermonde_pH=Dendermonde_pH.rename(columns={'Value':'PH','#Timestamp':'Date2'})
Dendermonde_pH=Dendermonde_pH.drop(columns={'Quality Code','Absolute Value','AV Quality Code'})
```

```python
Dendermonde_Watertemperature=Dendermonde_Watertemperature.rename(columns={'Value':'Watertemperature','#Timestamp':'Date'})
Dendermonde_Watertemperature=Dendermonde_Watertemperature.drop(columns={'Quality Code','Absolute Value','AV Quality Code'})
```

```python
Dendermonde_Turbidity=Dendermonde_Turbidity.rename(columns={'Value':'Turbidity','#Timestamp':'Date'})
Dendermonde_Turbidity=Dendermonde_Turbidity.drop(columns={'Quality Code','Absolute Value','AV Quality Code'})
```

```python
Dendermonde_Dissolved_Oxygen=Dendermonde_Dissolved_Oxygen.rename(columns={'Value':'Dissolved_Oxygen','#Timestamp':'Date'})
Dendermonde_Dissolved_Oxygen=Dendermonde_Dissolved_Oxygen.drop(columns={'Quality Code','Absolute Value','AV Quality Code'})
```

```python
Dendermonde_Conductivity=Dendermonde_Conductivity.rename(columns={'Value':'Conductivity','#Timestamp':'Date'})
Dendermonde_Conductivity=Dendermonde_Conductivity.drop(columns={'Quality Code','Absolute Value','AV Quality Code'})
```

## Preprocess In Order To Get Feature's Mean Value Per Day

```python
Dendermonde_pH['Date2']=Dendermonde_pH['Date2'].str[:10]
Dendermonde_pH=Dendermonde_pH.apply(lambda x: x.str.replace(',','.'))
Dendermonde_pH['PH']=Dendermonde_pH['PH'].astype(float).round(3)
Dendermonde_pH['Date2']=pd.to_datetime(Dendermonde_pH['Date2'])
Dendermonde_pH = Dendermonde_pH.set_index('Date2')
ph=Dendermonde_pH.reset_index().groupby(pd.Grouper(key='Date2', freq='D'))['PH'].mean()
ph=ph.reset_index()
ph=ph.rename(columns={'PH':'Mean PH'})
```

```python
Dendermonde_Watertemperature['Date']=Dendermonde_Watertemperature['Date'].str[:10]
Dendermonde_Watertemperature=Dendermonde_Watertemperature.apply(lambda x: x.str.replace(',','.'))
Dendermonde_Watertemperature['Watertemperature']=Dendermonde_Watertemperature['Watertemperature'].astype(float).round(3)
Dendermonde_Watertemperature['Date']=pd.to_datetime(Dendermonde_Watertemperature['Date'])
Dendermonde_Watertemperature = Dendermonde_Watertemperature.set_index('Date')
wt=Dendermonde_Watertemperature.reset_index().groupby(pd.Grouper(key='Date', freq='D'))['Watertemperature'].mean()
wt=wt.reset_index()
wt=wt.rename(columns={'Watertemperature':'Mean Watertemperature'})
```

```python
Dendermonde_Turbidity['Date']=Dendermonde_Turbidity['Date'].str[:10]
Dendermonde_Turbidity['Date']=pd.to_datetime(Dendermonde_Turbidity['Date'])
Dendermonde_Turbidity = Dendermonde_Turbidity.set_index('Date')
tur=Dendermonde_Turbidity.reset_index().groupby(pd.Grouper(key='Date', freq='D'))['Turbidity'].mean()
tur=tur.reset_index()
tur=tur.rename(columns={'Turbidity':'Mean Turbidity'})
```

```python
Dendermonde_Dissolved_Oxygen['Date']=Dendermonde_Dissolved_Oxygen['Date'].str[:10]
Dendermonde_Dissolved_Oxygen=Dendermonde_Dissolved_Oxygen.apply(lambda x: x.str.replace(',','.'))
Dendermonde_Dissolved_Oxygen['Dissolved_Oxygen']=Dendermonde_Dissolved_Oxygen['Dissolved_Oxygen'].astype(float).round(3)
Dendermonde_Dissolved_Oxygen['Date']=pd.to_datetime(Dendermonde_Dissolved_Oxygen['Date'])
Dendermonde_Dissolved_Oxygen = Dendermonde_Dissolved_Oxygen.set_index('Date')
do=Dendermonde_Dissolved_Oxygen.reset_index().groupby(pd.Grouper(key='Date', freq='D'))['Dissolved_Oxygen'].mean()
do=do.reset_index()
do=do.rename(columns={'Dissolved_Oxygen':'Mean Dissolved_Oxygen'})
```

```python
Dendermonde_Conductivity['Date']=Dendermonde_Conductivity['Date'].str[:10]
Dendermonde_Conductivity['Date']=pd.to_datetime(Dendermonde_Conductivity['Date'])
Dendermonde_Conductivity = Dendermonde_Conductivity.set_index('Date')
con=Dendermonde_Conductivity.reset_index().groupby(pd.Grouper(key='Date', freq='D'))['Conductivity'].mean()
con=con.reset_index()
con=con.rename(columns={'Dendermonde_Conductivity':'Mean Conductivity'})
```

## Concat in a singe Datagrame

```python
merged=pd.concat([wt,ph,tur,do,con],axis=1)
merged = merged.loc[:,~merged.T.duplicated(keep='first')]
merged=merged.drop(columns={'Date2'})
merged['Mean PH']=merged['Mean PH'].fillna(merged['Mean PH'].mean())
```

```python
merged
```

```python
merged.to_excel('Dendermonde_water_info_website_mean_per_day.xlsx',header=True)
```

## Approach without taking in consideration the day (keep multiple logs per day)

## Read Data 

```python
Dendermonde_pH=pd.read_csv('Dendermonde SF Dender_pH.csv',delimiter=';',header=7)
Dendermonde_Watertemperature=pd.read_csv('Dendermonde SF Dender_Watertemperature.csv',delimiter=';',header=7)
Dendermonde_Turbidity=pd.read_csv('Dendermonde SF Dender_Turbidity.csv',delimiter=';',header=7)
Dendermonde_Dissolved_Oxygen=pd.read_csv('Dendermonde SF Dender_Dissolved Oxygen.csv',delimiter=';',header=7)
Dendermonde_Conductivity=pd.read_csv('Dendermonde SF Dender_Conductivity.csv',delimiter=';',header=7)
```

## Preprocess

```python
Dendermonde_pH=Dendermonde_pH.rename(columns={'Value':'PH','#Timestamp':'Date2'})
Dendermonde_pH=Dendermonde_pH.drop(columns={'Quality Code','Absolute Value','AV Quality Code'})
Dendermonde_Watertemperature=Dendermonde_Watertemperature.rename(columns={'Value':'Watertemperature','#Timestamp':'Date'})
Dendermonde_Watertemperature=Dendermonde_Watertemperature.drop(columns={'Quality Code','Absolute Value','AV Quality Code'})
Dendermonde_Turbidity=Dendermonde_Turbidity.rename(columns={'Value':'Turbidity','#Timestamp':'Date3'})
Dendermonde_Turbidity=Dendermonde_Turbidity.drop(columns={'Quality Code','Absolute Value','AV Quality Code'})
Dendermonde_Dissolved_Oxygen=Dendermonde_Dissolved_Oxygen.rename(columns={'Value':'Dissolved_Oxygen','#Timestamp':'Date4'})
Dendermonde_Dissolved_Oxygen=Dendermonde_Dissolved_Oxygen.drop(columns={'Quality Code','Absolute Value','AV Quality Code'})
Dendermonde_Conductivity=Dendermonde_Conductivity.rename(columns={'Value':'Conductivity','#Timestamp':'Date5'})
Dendermonde_Conductivity=Dendermonde_Conductivity.drop(columns={'Quality Code','Absolute Value','AV Quality Code'})
```

```python
Dendermonde_pH=Dendermonde_pH.apply(lambda x: x.str.replace(',','.'))
Dendermonde_Watertemperature=Dendermonde_Watertemperature.apply(lambda x: x.str.replace(',','.'))
Dendermonde_Dissolved_Oxygen=Dendermonde_Dissolved_Oxygen.apply(lambda x: x.str.replace(',','.'))
Dendermonde_pH['PH']=Dendermonde_pH['PH'].astype(float).round(3)
Dendermonde_Watertemperature['Watertemperature']=Dendermonde_Watertemperature['Watertemperature'].astype(float).round(3)
Dendermonde_Dissolved_Oxygen['Dissolved_Oxygen']=Dendermonde_Dissolved_Oxygen['Dissolved_Oxygen'].astype(float).round(3)
```

```python
Dendermonde_pH['Date2']=Dendermonde_pH['Date2'].str[:19]
Dendermonde_pH['Date2']=pd.to_datetime(Dendermonde_pH['Date2'])

Dendermonde_Watertemperature['Date']=Dendermonde_Watertemperature['Date'].str[:19]
Dendermonde_Watertemperature['Date']=pd.to_datetime(Dendermonde_Watertemperature['Date'])

Dendermonde_Dissolved_Oxygen['Date4']=Dendermonde_Dissolved_Oxygen['Date4'].str[:19]
Dendermonde_Dissolved_Oxygen['Date4']=pd.to_datetime(Dendermonde_Dissolved_Oxygen['Date4'])

Dendermonde_Turbidity['Date3']=Dendermonde_Turbidity['Date3'].str[:19]
Dendermonde_Turbidity['Date3']=pd.to_datetime(Dendermonde_Turbidity['Date3'])

Dendermonde_Conductivity['Date5']=Dendermonde_Conductivity['Date5'].str[:19]
Dendermonde_Conductivity['Date5']=pd.to_datetime(Dendermonde_Conductivity['Date5'])
```

```python
Dendermonde_pH=Dendermonde_pH.drop_duplicates(subset=['Date2'])
Dendermonde_Watertemperature=Dendermonde_Watertemperature.drop_duplicates(subset=['Date'])
Dendermonde_Dissolved_Oxygen=Dendermonde_Dissolved_Oxygen.drop_duplicates(subset=['Date4'])
Dendermonde_Turbidity=Dendermonde_Turbidity.drop_duplicates(subset=['Date3'])
Dendermonde_Conductivity=Dendermonde_Conductivity.drop_duplicates(subset=['Date5'])
```

## Join  Dataframes Based on Date in order to be specific (down to seconds)

```python
m1=pd.merge(Dendermonde_Watertemperature,Dendermonde_pH,left_on='Date',right_on='Date2',how='left')
m2=pd.merge(Dendermonde_Turbidity,Dendermonde_Dissolved_Oxygen,left_on='Date3',right_on='Date4',how='left')
m3=pd.merge(m1,m2,left_on='Date',right_on='Date3',how='inner')
final_df_daily=pd.merge(m3,Dendermonde_Conductivity,left_on='Date',right_on='Date5',how='left')
```

```python
final_df_daily
```

## Preprocess Again

```python


final_df_daily=final_df_daily.drop(columns={'Date2','Date3','Date4','Date5'})

final_df_daily['PH']=final_df_daily['PH'].fillna(final_df_daily['PH'].mean())
final_df_daily['Conductivity']=final_df_daily['Conductivity'].fillna(final_df_daily['Conductivity'].mean())
final_df_daily['PH']=final_df_daily['PH'].round(1)
final_df_daily['Conductivity']=final_df_daily['Conductivity'].round(1)
```
