```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={'SARIMAX': [-3.2],'MLP': [0.973],"LSTM (next day's prediction)": [0.0021],"LSTM (next hour's prediction)": [0.8906]}
```

```python
eval=pd.DataFrame(data=data)
```

```python
eval_t=eval.T
```

```python
eval_t=eval_t.rename(columns={0:'R2'})
```

```python
eval_t['MSE']=[39243.49,214.329,8390.180,933.01]
```

```python
eval_t.sort_values(by=['R2'], ascending=False,inplace=True)
```

```python
eval_t.sort_values(by=['R2'], ascending=False,inplace=True)
```

```python
plt.figure(figsize=(25, 10))
plt.barh(eval_t.index, eval_t['R2'],color=(0.2, 0.4, 0.6, 0.6),edgecolor='black')
plt.axvline(x=0.0, color='black', linestyle='-', linewidth=3)
plt.title("Coefficient of Determination (R2) per Model",fontsize=20,)
plt.xlabel("R2",fontsize=20)
plt.ylabel("Model",fontsize=20)
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)
for index, value in enumerate(eval_t['R2']):
    plt.text(value, index,
             str(value),fontsize=20)

#plt.savefig('r2_per_model.png')
plt.show()
```

```python
plt.figure(figsize=(25, 10))
ax=eval_t['MSE'].plot(kind='bar',color='seagreen',edgecolor='black',rot=30)
plt.title("Mean Squared Error (MSE) per Model",fontsize=20)
plt.xlabel("Model",fontsize=20)
plt.ylabel("MSE",fontsize=20)
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005),fontsize=20)
plt.legend(fontsize=20)
#plt.savefig('mse_per_model.png')
plt.show()
```
